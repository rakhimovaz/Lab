```python
from PIL import Image

# Open the image file
img = Image.open("C:/Users/Zuhra/Downloads/Piona flower.jpg")

# Print image properties
print(img.size)  # width and height in pixels
print(img.mode)  # color mode (RGB, L, etc.)

# Display image
img
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Read image using matplotlib
img_mpl = mpimg.imread("C:/Users/Zuhra/Downloads/Piona flower.jpg")

# Print shape and data type
print("Image shape:", img_mpl.shape)
print("Data type:", img_mpl.dtype)

# Display the image
plt.imshow(img_mpl)
plt.axis("off")
plt.show()
```

```python
from skimage import io, img_as_float

# Read image with scikit-image
img_sk = io.imread("C:/Users/Zuhra/Downloads/Piona flower.jpg")
print("Original dtype:", img_sk.dtype)
print("Original range:", img_sk.min(), "-", img_sk.max())

# Convert to float format (0-1 range)
img_float = img_as_float(img_sk)

# Print min and max pixel values
print("Float dtype:", img_float.dtype)
print("Min pixel value:", img_float.min())
print("Max pixel value:", img_float.max())

# Display
plt.imshow(img_float)
plt.axis("off")
plt.show()
```

```python
from skimage import io, img_as_float
import matplotlib.pyplot as plt

# Read and convert image
img_sk = io.imread("C:/Users/Zuhra/Downloads/Piona flower.jpg")
img_float = img_as_float(img_sk)

# Print pixel value range
print(img_float.min(), img_float.max())

# Display with colorbar
plt.imshow(img_float)
plt.axis("off")
plt.colorbar()
plt.show()
```

```python
import cv2

gray = cv2.imread("\\Users\\Zuhra\\Downloads\\Piona flower.jpg", cv2.IMREAD_GRAYSCALE)
bgr = cv2.imread("\\Users\\Zuhra\\Downloads\\Piona flower.jpg")
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 4))   # BONUS

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.show()
```

```python
import cv2

# Read image in grayscale
gray = cv2.imread("C:/Users/Zuhra/Downloads/Piona flower.jpg", cv2.IMREAD_GRAYSCALE)

# Read image in color (BGR format)
bgr = cv2.imread("C:/Users/Zuhra/Downloads/Piona flower.jpg")

# Convert BGR to RGB for proper display
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Display both grayscale and RGB images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.show()
```
